```
import numpy as np
embeddings_index ={}
with open('glove.6B.50d.txt',encoding='utf-8') as f:
  for line in f:
    values = line.split()
    word = values[0]
    coefs = np.asarray(values[1:],dtype='float')
    embeddings_index[word] = coefs
  f.close()
```

```
embeddings_index["eat"]
```

```
def embedding_output(X):
  embedding_matrix_output = np.zeros((X.shape[0],10,50))
  for ix in range(X.shape[0]):
      for jx in range(len(X[ix])):
          embedding_matrix_output[ix][jx] = embeddings_index[X[ix][jx].lower()]
  return embedding_matrix_output
  
embedding_matrix_train  = embedding_output(X_train)
embedding_matrix_test  = embedding_output(X_test)
```

```
print(embedding_matrix_train.shape)
```

***Step 5 Define the RNN/LSTM Model***

```
from keras.models import Sequential
from keras.layers import *
```

```
model = Sequential()
model.add(LSTM(64,input_shape=(10,50),return_sequences=True))
model.add(Dropout(0.4)) #regularization
model.add(LSTM(64,input_shape=(10,50))) #stacking
model.add(Dropout(0.3)) #regularization
model.add(Dense(5))
model.add(Activation('softmax'))
model.summary()
```

```
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```
hist = model.fit(embedding_matrix_train,Y_train,epochs=150,batch_size=64,shuffle=True,validation_split=0.1)
```

```
model.evaluate(embedding_matrix_test,Y_test)
```

```
pred = model.predict_classes(embedding_matrix_test)
print(pred)
```
